```python
import pandas as pd
import numpy as np
```

```python
# target area: x=20..30, y=-10..-5
tar_x = (20,30)
tar_y = (-10,-5)
```

```python
# target area: 70..96, y=-179..-124
tar_x = (70,96)
tar_y = (-179,-124)
```

```python
def x_step(position, vel):
    # position, vel = pair
    position += vel
    if vel > 0:
        vel = vel - 1 
    if vel < 0:
        vel = vel + 1 
    return position, vel

def y_step(position, vel):
    # position, vel = pair
    return position + vel, vel-1

# x_step(7,6)
```

```python
possible_x = {}
possible_y = {}
for x_velocity in range(1,501):
    # generate x steps
    x_steps = []
    pos = 0
    x_vel = x_velocity
    while x_vel > 0:
        pos,x_vel = x_step(pos,x_vel)
        x_steps.append(pos)
    possible_x[x_velocity] = x_steps

for y_velocity in range(-501,501):
    y_steps = []
    pos = 0
    y_vel = y_velocity
    while pos > min(tar_y):
        pos,y_vel = y_step(pos,y_vel)
        y_steps.append(pos)
    possible_y[y_velocity] = y_steps

# possible_x, possible_y
```

```python
true_x = {}
true_y = {}
for pos in possible_x:
    if any([i >= min(tar_x) and i <= max(tar_x) for i in possible_x[pos]]):
        true_x[pos] = possible_x[pos]
for pos in possible_y:
    if any([i >= min(tar_y) and i <= max(tar_y) for i in possible_y[pos]]):
        true_y[pos] = possible_y[pos]

# true_x, true_y
```

```python
possible_coor = []
for pos in true_x:
    for idx in range(len(true_x[pos])):
        if true_x[pos][idx] >= min(tar_x) and true_x[pos][idx] <= max(tar_x):
            for pos_y in true_y:
                if idx < len(true_x[pos])-1:
                    if idx < len(true_y[pos_y]) and true_y[pos_y][idx] >= min(tar_y) and true_y[pos_y][idx] <= max(tar_y):
                        possible_coor.append((pos,pos_y))
                elif idx == len(true_x[pos])-1:
                    #x is static - the probe is falling down
                    for idy in range(idx, len(true_y[pos_y])):
                        if true_y[pos_y][idy] >= min(tar_y) and true_y[pos_y][idy] <= max(tar_y):
                            possible_coor.append((pos,pos_y))

sorted(possible_coor, key= lambda x: x[1], reverse=True)
```
