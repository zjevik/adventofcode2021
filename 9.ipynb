```python
arr_over = {}
for level in range (0,10):
    current = np.zeros(arr.shape)
    for y in range(arr.shape[0]):
        for x in range(arr.shape[1]):
            if arr[y,x] > level:
                current[y,x] = 1
    arr_over[level] = current
# arr_over
```

```python
def get_neigh_coor(coor, arr):
    ret = []
    for i in range(-1,2):
        for j in range(-1,2):
            if i != 0 or j != 0:
                ret.append((coor[0]+i,coor[1]+j))
    return [x for x in ret if x[0]>=0 and x[1]>=0 and x[0] < arr.shape[0] and x[1] < arr.shape[1]]
def get_ortog_neigh_coor(coor, arr):
    ret = []
    for i in range(-1,2):
        j=0
        if i != 0 or j != 0:
            ret.append((coor[0]+i,coor[1]+j))
    for j in range(-1,2):
        i = 0
        if i != 0 or j != 0:
            ret.append((coor[0]+i,coor[1]+j))
    return [x for x in ret if x[0]>=0 and x[1]>=0 and x[0] < arr.shape[0] and x[1] < arr.shape[1]]
```

```python
# part 1
ans = 0
for y in range(arr.shape[0]):
    for x in range(arr.shape[1]):
        if np.array([arr_over[arr[y,x]][nb[0],nb[1]] for nb in get_neigh_coor((y,x),arr)]).mean() == 1:
            #all neighbors are bigger
            ans += 1+arr[y,x]
ans
        
```
