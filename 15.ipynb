```python
def get_version_and_type(bits):
    # print(bits[0:3])
    version = int(bits[0:3],2)
    bits = bits[3:]
    # print(bits[0:3])
    type = int(bits[0:3],2)
    print(f'version: {version}, type: {type}')
    return version, type, bits[3:]

def get_literal(bits):
    result = ''
    while bits[0] == '1':
        result += bits[1:5]
        bits = bits[5:]
    result += bits[1:5]
    bits = bits[5:]
    return int(result,2), bits
def get_operator(bits):
    len_type_id = int(bits[0])
    bits = bits[1:]
    if len_type_id == 0:
        total_length = int(bits[:15],2)
        print(f'total_len: {total_length}. Parsing: [{bits[:total_length]}]')
        bits = bits[15:]
        packs = full_parse(bits[:total_length])
        bits = bits[total_length:]
        return len_type_id,bits,-1,total_length,packs
    elif len_type_id == 1:
        num_of_subpackets = int(bits[:11],2)
        print(f'num_of_subpackets: {num_of_subpackets}')
        bits = bits[11:]
        bits,packs = num_parse(bits,num_of_subpackets)
        return len_type_id,bits,num_of_subpackets,-1,packs

def full_parse(bits):
    packets = []
    bit_len = len(bits)-1
    while bit_len != len(bits) and len(bits)>0 and int(bits,2)>0:
        bit_len = len(bits)
        print(bits)
        version,type,bits = get_version_and_type(bits)
        # print(f'version,type,bits, {(version,type,bits)}')
        if type == 4:
            #literal
            lit,bits = get_literal(bits)
            packets.append((type,lit))
            # print(f'literal: {(lit,bits)}')
        else:
            #operator
            len_type_id,bits,num_of_subpackets,str_len,packs = get_operator(bits)
            packets.append((type,num_of_subpackets,str_len,packs))
            # if len_type_id == 1:
                # print(f'number of subpackets: {num_of_subpackets}')
    return packets

def num_parse(bits,n):
    packets = []
    bit_len = len(bits)-1
    for i in range(n):
        bit_len = len(bits)
        version,type,bits = get_version_and_type(bits)
        # print(f'version,type,bits, {(version,type,bits)}')
        if type == 4:
            #literal
            lit,bits = get_literal(bits)
            packets.append((type,lit))
            # print(f'literal: {(lit,bits)}')
        else:
            #operator
            len_type_id,bits,num_of_subpackets,str_len,packs = get_operator(bits)
            packets.append((type,num_of_subpackets,str_len,packs))
            # if len_type_id == 1:
                # print(f'number of subpackets: {num_of_subpackets}')
    return bits, packets
```
